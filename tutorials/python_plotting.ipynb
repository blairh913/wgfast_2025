# Plotting

```python
import pandas as pd
```

Read csv file

```python
nfs = pd.read_csv("nearfield.csv")
nfs
```

See summary

```python
nfs.describe()
```

Set Index

```python
nfs = nfs.set_index(["frequency", "bw"])
nfs
```

Plot all points

```python
nfs.reset_index().plot.scatter(x="frequency", y="bw", #x and y axes
                               c="rnf", #color vairable
                               clim=[0,10], #colorscale limits
                               s=2, #size of scatter
                               xlabel="Frequency (Hz)", #x axis title
                               ylabel="Beam width (°)") #y axis title
```

Query the dataframe to plot only lines for 18, 38, 70 kHz and beam widths 4,7,18 ° 

```python
nfs.query('frequency in [18000, 38000, 70000] & bw in [4,7,18]' ).rnf.unstack().plot(title="Nearfield (m)", 
                                                                                             ylabel="Nearfield (m)",
                                                                                             xlabel="Frequency (Hz)",
                                                                                             xticks=[18000,38000,70000])
```
